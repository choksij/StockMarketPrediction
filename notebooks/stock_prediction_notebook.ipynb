```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv("C:/Users/choks/stock_market_prediction/data/processed/processed_stock_data.csv", index_col='Date', parse_dates=True)
```

```python

# Plot the closing price
plt.figure(figsize=(14, 7))
plt.plot(data['Close'])
plt.title("Stock Closing Price")
plt.show()
```

```python

# Plot moving averages
data['20_SMA'] = data['Close'].rolling(window=20).mean()
data['50_SMA'] = data['Close'].rolling(window=50).mean()
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close')
plt.plot(data['20_SMA'], label='20 Day SMA')
plt.plot(data['50_SMA'], label='50 Day SMA')
plt.legend()
plt.show()
```
